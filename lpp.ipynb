```python
from utils import *
def lpp(X,t,d_):
    '''
    Locality Preserving Projection \n
    X: (d,n) matrix
    '''
    W = np.exp(1/t*dist_2m_sq(X.T,X.T))
    D = np.diag(W.sum(1))
    L  = D-W
    lhs = X.dot(L).dot(X.T)
    rhs = X.dot(D).dot(X.T)
    cov = np.linalg.inv(rhs).dot(lhs)
    eigval, eigvec = np.linalg.eig(cov)
    idx = np.argpartition(eigval, -d_)[-d_:]
    selected_vec = eigvec[:,idx]
    return selected_vec.T.dot(X)
lpp(np.random.rand(20,100),0.1,10).shape
dist_2m_sq(np.random.rand(10,10),np.random.rand(10,10))
```
